### 2.1 So I have created a new column (new_column) and assigned the values of that column to be NAN(Not a number) 

```python
df['new_column'] = np.nan #np is alias of numpy
df.head()
df['new_column'].isna().sum() #Displays the count of rows where new_column is NAN
```

###  2.2 I have defined a function(add_a_number) which takes 3 parameters label, starting_number, ending_number. This function replaces the NAN value of new_column within a specific index range.

```python
def add_a_number(label , starting_number , ending_number):
    df.loc[starting_number : ending_number,'new_column'] = label
```

### 2.3 Given in question that it has to keep on asking till all rows are labelled so i have used a while loop to check whether the NAN count is 0 or not. If not then it keeps on asking label, starting number, ending number till all the rows are replaced by a label.

```python
while (df['new_column'].isna().sum()!=0):
    print("enter label")
    label = int(input())
    print("starting_number")
    starting_number = int(input())
    print("ending_number")
    ending_number = int(input())
    add_a_number(label,starting_number,ending_number)
```
